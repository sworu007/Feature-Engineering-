```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'/content/sample_data/Social_Network_Ads.csv')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop(['Purchased','User ID','Gender'], axis=1), df['Purchased'], test_size=0.2, random_state=0)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x_train)
x_train_sc = sc.transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
sc.mean_
```

```python
x_train_sc[:5]
```

```python
x_train_sc = pd.DataFrame(x_train_sc, columns=x_train.columns)
x_test_sc = pd.DataFrame(x_test_sc, columns=x_test.columns)
```

```python
x_train_sc[:5]
```

```python
x_train_sc.describe()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr_sc = LogisticRegression()

lr.fit(x_train, y_train)
lr_sc.fit(x_train_sc, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred_sc = lr_sc.predict(x_test_sc)
```

```python
from sklearn.metrics import accuracy_score
print(("Without STD +"),accuracy_score(y_test, y_pred))
print(("After STD ="),accuracy_score(y_test, y_pred_sc))
```
